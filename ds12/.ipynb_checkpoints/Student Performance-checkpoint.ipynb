```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("student_performance.csv")
df.head()
```

```python
students=df['Student_ID'].count()
print("total student registered =",students)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
average_final_score=df['Final_Score'].mean()
print("Average final score",average_final_score)
```

```python
minimum_score=df['Final_Score'].min()
print("minimum score is =",minimum_score)
```

```python
average_attendence=df['Attendance'].mean()
print("average attendence",average_attendence)
```

```python
average_final_score_by_gender=df.groupby('Gender')['Final_Score'].mean()
print("average final score by gender",average_final_score_by_gender)
```

```python
avg_final_score_by_parentedu=df.groupby('Parental_Education')['Final_Score'].mean()
print(avg_final_score_by_parentedu)
```

```python
df['Gender'].value_counts()
```

```python
df['Internet_Access'].value_counts()
```

```python
avg_score_internet_access=df.groupby('Internet_Access')['Final_Score'].mean()
print("average score with having internet ",avg_score_internet_access)
```

```python
total_study_gender=df.groupby('Gender')['Study_Hours'].mean()
print("total study hours gender wise",total_study_gender)
```

```python
top_75=(df["Final_Score"]>=75).mean()*100
print("student scored above 75 percent=",top_75)
```

```python
top_50=(df['Final_Score']>50).mean()*100
print("students scoring grater than 50 percent",top_50)
```

```python
failure=(df['Final_Score']<=33).mean()*100
print("faliure rate",failure)
```
