```python
# step 1:- Importion Required libraries for EDA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# step 2,3:- load the dataset
df=pd.read_csv("data.csv")
df.head()
```

```python
# step 4:- Display last 5 rows
df.tail()
```

```python
# step 5:- count rows and columns
r, c=df.shape
print("Rows: ",r)
print("Columns: ",c)
```

```python
# step 6:- Checking the type of data
df.dtypes
```

```python
# step 7:- display column names
df.columns
```

```python
# step 8:- Drop the irrelevent coluns
df=df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
```

```python
df.head()
```

```python
# Renaming the columns
df=df.rename(columns={
    'Engine HP': 'HP',
    'Engine Cylinders': 'Cylinders',
    'Transmission Type': 'Transmission',
    'Driven_Wheels': 'Drive Mode',
    'highway MPG': 'MPG-H',
    'city mpg': 'MPG-C',
    'MSRP': 'Price'
})
df.head()
```

```python
# Droping the duplicate rows
duplicate_rows_df=df[df.duplicated()]
print("Number of duplicate rows: ",duplicate_rows_df.shape)
```

```python
df=df.drop_duplicates()
df.shape
```

```python
# Droping the missing or null values
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
# Box Plot:- it is used for value distribution
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
#make a plot to represent make no. of cars
no_of_cars=df.Make.value_counts()
print(no_of_cars)
```

```python
#visualize data
df.Make.value_counts().nlargest(20).plot(kind='bar')
plt.title("number of cars made by top company")
plt.xlabel("COMPANY")
plt.ylabel("number of cars")
plt.show()
```

```python
#scatter plot between HP and price
plt.scatter(df['HP'],df['Price'])
plt.title("relation between hoursepower and price")
plt.xlabel("HP")
plt.ylabel('Price')
plt.show()
```
