```python
#import essentials variables
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df=sns.load_dataset("tips")
df.head()
```

```python
# scatter plot: it is used to show realtion ship b/w two variable
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.title("Scatter plot")
plt.show()
```

```python
#line plot: it is used to show trend visualization
sns.lineplot(x="size",y="total_bill",data=df)
plt.title("line plot")
plt.show()
```

```python
#bar plot: bar plot is used to show avg. value
sns.barplot(x="day",y="total_bill",data=df)
plt.title("bar plot")
plt.show()
```

```python
#count plot: it count the frequency of catogories
sns.countplot(x="day",data=df)
plt.title("count plot")
plt.show()
```

```python
#histogram plot: shows distribution of data
sns.histplot(df["total_bill"],bins=10)
plt.show()
```
