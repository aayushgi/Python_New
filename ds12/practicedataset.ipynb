# Titanic dataset

```python
#import essentials variables
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
                    
```

```python
#1: count total passenger
total=len(df)
print("total no. of passenger",total)
```

```python
#2: how many male female survived
a=df[df["alive"]=="yes"]
print("total female survived",len(a))
```

```python
#3 count missing values in each column
df.isnull().sum()
```

```python
#4 create countlpot of survived column
sns.countplot(x="survived",data=df)
plt.show()
```

```python
#5 craete barplot of survival rate by class
sns.barplot(x="class",y="survived",data=df)
plt.show()
```
