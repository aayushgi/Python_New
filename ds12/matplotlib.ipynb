# Matplotlib

# line plot

```python
import matplotlib.pyplot as plt
# line plot
x=[1,2,3,4]
y=[10,12,29,40]
plt.plot(x,y)
plt.show()
```

# adding lables, title and grid

```python
x=[1,2,3,4]
y=[10,12,29,40]
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("frist plot")
plt.grid(True)
plt.show()
```

# Styling

```python
plt.plot(x,y,color="blue",linestyle="--",marker="o")
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y,color="green",linestyle=":",marker="^")
```

```python
plt.plot(x,y,color="red",linestyle="-",marker="s")
```

# Multiple line

```python
y2=[5,15,20,35]
plt.plot(x,y,label="line A")
plt.plot(x,y2,label="line B")
plt.legend()
plt.show()
```

# Bar chart

```python
names=["A","B","C"]
values=[10,24,36]
plt.bar(names,values)
plt.title("bar chart example")
plt.show()
```

# Scatter plot

```python
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

# Histogram

```python
data=[1,2,2,3,3,3,4,4,5]
plt.hist(data,bins=5,color="orange")
plt.title("histogram")
plt.show()
```

# Saving

```python
plt.plot(x,y)
plt.savefig("my_plot.png")
plt.show()
```
