# Matplotlib

# line plot

```python
import matplotlib.pyplot as plt
# line plot
x=[1,2,3,4]
y=[10,12,29,40]
plt.plot(x,y)
plt.show()
```

# adding lables, title and grid

```python
x=[1,2,3,4]
y=[10,12,29,40]
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("frist plot")
plt.grid(True)
plt.show()
```

# Styling

```python
plt.plot(x,y,color="blue",linestyle="--",marker="o")
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y,color="green",linestyle=":",marker="^")
```

```python
plt.plot(x,y,color="red",linestyle="-",marker="s")
```

# Multiple line

```python
y2=[5,15,20,35]
plt.plot(x,y,label="line A")
plt.plot(x,y2,label="line B")
plt.legend()
plt.show()
```

# Bar chart

```python
names=["A","B","C"]
values=[10,24,36]
plt.bar(names,values)
plt.title("bar chart example")
plt.show()
```

# Scatter plot

```python
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```

# Histogram

```python
data=[1,2,2,3,3,3,4,4,5]
plt.hist(data,bins=5,color="orange")
plt.title("histogram")
plt.show()
```

# Saving

```python
plt.plot(x,y)
plt.savefig("my_plot.png")
plt.show()
```

# Tasks

```python
#1=create simple line plot
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
```

```python
#2=adding x lable and y lable on a line plot
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
#3=change line to "--" & marker as "o"
plt.plot(x,y,linestyle="--",marker="o")

plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
#4= plot 2 line on same graph
#y1=x
#y2=x^2
x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[1,4,9,16,25]
plt.plot(x,y1,label="line A")
plt.plot(x,y2,label="line B")
plt.legend()
plt.show()
```

```python
#5=add a legend to distinguish multiple line
x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[1,4,9,16,25]
plt.plot(x,y1,label="line A")
plt.plot(x,y2,label="line B")
plt.legend()
plt.show()
```

```python
#6=save plot as png

x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[1,4,9,16,25]
plt.plot(x,y1,label="line A")
plt.plot(x,y2,label="line B")
plt.legend()
plt.savefig("pltted graph")
plt.show()
```

```python
#7create a bar chart
students=["A","B","C","D"]
marks=[85,75,92,88]          
plt.bar(students,marks)
plt.title("students chart")
plt.show()
```

```python
#10=add grid lines

x=[10,20,30,40,50]
y=[5,15,25,35,45]
plt.scatter(x,y,marker="s")
plt.grid(True)

plt.title("scatter plot")
plt.show()
```
