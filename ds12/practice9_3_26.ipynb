```python
data={
    "id":[1,2,3,4,5],
    "name":["amit","neha","raj","simran","arjun"],
    "department":["IT","HR","IT","Finance","IT"],
    "salary":[50000,45000,60000,55000,70000],
    "experience":[3,2,5,4,6],
    "city":["Delhi","Mumbai","Delhi","Pune","banglore"]
}
```

```python
# Create a DataFrame from the dataset. 
df=pd.DataFrame(data)
df
```

```python
#13 . Find maximum salary employee.
print(df["salary"].max())
```

```python
#14  Count number of employees city-wise. 
print(df[df["city"]=="Delhi"].())
```

```python
#15 sort employees by salary (descending). 
print(df.sort_values(by="salary",ascending=False))
```

```python
data3={
    "departments": ["IT","HR","Finance","Sales"] ,

    "employees" :[20,10,8,12] 
}
```

```python
df3=pd.DataFrame(data3)
df3
```

# line plot

```python
#Plot Experience vs Salary using a line chart. 

x=df2.experience
y=df2.salary
plt.xlabel("experience")
plt.ylabel("salary")
plt.plot(x,y,color="blue",marker="o")
```

```python
# Create a bar chart for number of employees per department. 
x=df3.employees
y=df3.departments
plt.bar(x,y)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=df2.experience
y=df2.salary
plt.scatter(x,y)
plt.title("scatter plot")
plt.show()
```
