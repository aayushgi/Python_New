```python
import numpy as np
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
```

```python
#input data
X=np.array([1,2,3,4],dtype=float)
```

```python
#output
Y=np.array([2,4,6,8],dtype=float)
```

```python
#create ANN model
model=tf.keras.Sequential([
    #one neuron in dense layer
    tf.keras.layers.Dense(units=1,input_shape=[1])
])
```

```python
#compile model
model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)
```

```python
model.fit(X,Y,epochs=500)
```

```python
res=model.predict(np.array([190],dtype=float))[0][0].round()
print("result",res)
```
